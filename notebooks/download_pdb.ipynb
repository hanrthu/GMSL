```python
import csv
import codecs
from tqdm import tqdm
from urllib.request import urlretrieve
```

```python
rows = []
with codecs.open('../data/EC/pdb_chain_enzyme.csv', encoding='utf-8') as f:
    for row in csv.DictReader(f, skipinitialspace=True):
        rows.append(row)
print(len(rows))
```

```python
pdbids = [row['PDB'] for row in rows]
```

```python
pdbids = list(set(pdbids))
```

```python
len(pdbids)
```

```python
failed = 0
root_dir = '../data/EC/'
idx = 0
download_pdb = []
failed_pdb = []
for pdbid in tqdm(pdbids):
    
    # if idx == 10:
    #     break
    try:
        url = 'https://files.rcsb.org/download/' + pdbid.lower() + '.pdb'
        # print(url)
        urlretrieve(url, root_dir +'all/' + pdbid.upper() + '.pdb')
        download_pdb.append(pdbid.lower())
    except:
        print("cannot download:", pdbid)
        failed += 1
        failed_pdb.append(pdbid.lower())
    idx += 1
```

```python
print(download_pdb)
print(failed_pdb)
```

```python
pdb_dic = {}
for row in rows:
    if row['PDB'] in pdb_dic.keys():
        pdb_dic[row['PDB']].append(row['EC_NUMBER'])
    else:
        pdb_dic[row['PDB']] = [row['EC_NUMBER']]
```

```python
for key, value in pdb_dic.items():
    pdb_dic[key] = list(set(value))
for key, value in pdb_dic.items():
    if len(value) != 1:
        print(key)
```
